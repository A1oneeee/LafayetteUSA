````python
# importing the dependecies
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
````

Data Collection and Pre-Processing

````python
# load data
raw_mail_data = pd.read_csv('/content/mail_data.csv')
````

````python
print(raw_mail_data)
````

````python
raw_mail_data.isnull().sum()
````

````python
# replace the null value with a null string
mail_data = raw_mail_data.where((pd.notnull(raw_mail_data)),'')
````

````python
mail_data.head()
````

````python
# checking the number of rows and columns in the dataframe
mail_data.shape
````

Label Encoding

````python
# label spam mail as 0; ham as 1
mail_data.loc[mail_data['Category'] == 'spam', 'Category',] = 0
mail_data.loc[mail_data['Category'] == 'ham', 'Category',] = 1
````

````python
mail_data.head()
````

````python
# seperating the data as text and label
X = mail_data['Message']
Y = mail_data['Category']
````

````python
print(X)
````

````python
print(Y)
````

Spliting the data into Training Data and Test Data

````python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=3)
````

````python
print(X.shape)
print(X_train.shape)
print(X_test.shape)
````

````python
#transform the text data t ofeature vectors that can be used as input to the Logistic regression
feature_extraction = TfidfVectorizer(min_df=1, stop_words='english', lowercase=np.True_)
X_train_features = feature_extraction.fit_transform(X_train)
X_test_features = feature_extraction.transform(X_test)

# convert Y_train and Y_test values as integers
Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')
````

````python
print(X_train)
````

````python
print(X_train_features)
````

Training the Model

Logistic Regression

````python
model = LogisticRegression()
````

````python
# training the model with the training data
model.fit(X_train_features, Y_train)
````

Eveluating the trained Model

````python
# prediction on training data
prediction_on_training_data = model.predict(X_train_features)
accuracy_on_training_data = accuracy_score(Y_train, prediction_on_training_data)
````

````python
print('Accurency on training data : ', accuracy_on_training_data)
````

````python
# prediction on test data
prediction_on_training_data = model.predict(X_test_features)
accuracy_on_training_data = accuracy_score(Y_test, prediction_on_training_data)
````

````python
print('Accurency on test data : ', accuracy_on_training_data)
````

Building a Predictive System

````python
input_mail = ["2 a wkly comp to win FA Cu005. Tto receive entry question(std txt rate)T&C's apply 08452810075"]

# convert text to feature vectors
input_data_features = feature_extraction.transform(input_mail)

#making prediction
prediction = model.predict(input_data_features)
print(prediction)

if (prediction[0]==1):
  print('Ham mail')

else:
  print('Spam mail')
````
